##Extracci贸n de los datos .json

```python
url='https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
request=requests.get(url)
df=request.json()
datos=pd.json_normalize(df)
```

# Transformaci贸n

##Transformando datos numericos

**Aqui vamos a empezar con la columna account.Charges.Total que es la unica columna con datos numericos que necesita tranformase, la vamos a cambiar a tipo float y comprobaremos que no nos quede ningun valor nulo**

```python
datos['account.Charges.Total']=pd.to_numeric(datos['account.Charges.Total'],errors='coerce')
datos.info()
```

```python
datos.isnull().sum()
```

**Como obtuvimos valores nulos, los volveremos 0 y comprobaremos nuevamente**

```python
datos['account.Charges.Total']=datos['account.Charges.Total'].fillna(0)
datos.isnull().sum()
```
