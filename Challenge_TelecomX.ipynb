##ImportaciÃ³n de paqueterias

```python
import pandas as pd
import matplotlib as plt
import requests
import numpy as np
```

##ExtracciÃ³n de los datos .json

```python
url='https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'
request=requests.get(url)
df=request.json()
datos=pd.json_normalize(df)
```

```python
datos.sample(10)
```

```python
datos.info()
```

```python
datos.isnull().sum()
```

#ðŸ”§ TransformaciÃ³n

##Transformando datos numericos

**Aqui vamos a empezar con la columna account.Charges.Total que es la unica columna con datos numericos que necesita tranformase, la vamos a cambiar a tipo float y comprobaremos que no nos quede ningun valor nulo**

```python
datos['account.Charges.Total']=pd.to_numeric(datos['account.Charges.Total'],errors='coerce')
datos.info()
```

```python
datos.isnull().sum()
```

Convertimos los datos nulos a 0

```python
datos['account.Charges.Total']=datos['account.Charges.Total'].fillna(0)
datos['account.Charges.Total'].isnull().sum()
```

##Transformando datos booleanos

Evaluaremos las columnas que parecen tener valores de si y no para comprobar que no tengan otro tipo de dato, evaluaremos las columnas:

Churn,

```python
col_check=['Churn','customer.Partner','customer.Dependents','phone.PhoneService','phone.MultipleLines','internet.OnlineBackup',
'internet.DeviceProtection','internet.TechSupport','internet.StreamingTV','internet.StreamingMovies','account.PaperlessBilling']

for col in col_check:
  print(f'Columna {col}: {datos[col].unique()}')
```

Ahora que sabemos cuales son los que tienen solo valores de Si y NO los haremos booleanos, los que tengan otro dato lo haremos categorico.

```python
colum_bool=['Churn','customer.Partner','customer.Dependents','phone.PhoneService','account.PaperlessBilling']
datos[colum_bool]=datos[colum_bool].astype(np.bool_)
datos.info()
```

```python
datos.sample(5)
```

##Transformando datos string

Ahora solo queda transformar los datos que serÃ¡n texto

```python
#Crear lista para filtrar datos
colum_str=['customer.gender','phone.MultipleLines','internet.InternetService','internet.OnlineSecurity','internet.OnlineBackup','internet.DeviceProtection','internet.TechSupport','internet.StreamingTV',
'internet.StreamingMovies','account.Contract','account.PaymentMethod']

#Manipulamos los datos para que sean minisculas
for i in colum_str:
  datos[i]=datos[i].str.lower()
datos.sample(10)
```

Asi queda nuestra manipulacion de datos
